```python
wiki = pd.read_html('https://pt.wikipedia.org/wiki/Grammy_Award_para_%C3%A1lbum_do_ano')
len(wiki)
```

```python
len_wiki = len(wiki)

for n in range(2,len_wiki):
  if n == 2:
    wiki_consolidado = wiki[n]
    print('wiki_consolidado:', wiki_consolidado)
  else:
    wiki_consolidado = pd.concat([wiki_consolidado,wiki[n]])
wiki_consolidado
```
